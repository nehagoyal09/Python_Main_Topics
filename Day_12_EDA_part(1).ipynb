# **Handling Outliers Using IQR**

 **IQR** (Interquartile Range)

## **Quartiles**

- Q1 (25th percentile)  lower bound

- Q3 (75th percentile)  upper bound

- IQR = Q3  Q1

```python
values = [1, 2, 3, 4, 5, 6, 9, 10, 12, 55, 60]
print(values)
```

```python
Q1 = np.percentile(values, 25)
print(Q1)
```

```python
Q3 = np.percentile(values, 75)
print(Q3)
```

```python
IQR = Q3 - Q1
print(IQR)
```

## **Outliers Boundaries**

- Lower Bound = Q1  1.5  IQR

- Upper Bound = Q3 + 1.5  IQR

```python
low = 3.5 - 1.5 * 7.5
upp = 11 + 1.5 * 7.5

print(low, upp)
```

```python
outliers = []
```

```python
print(values)
```

```python
for i in values:
  if((i > upp) or (i < low)):
    outliers.append(i)
```

```python
outliers
```

```python
for i in values:
  if i not in outliers:
    print(i)
```

```python
Result = [i for i in values if i not in outliers]
print(Result)
```
