# **Measure of Dispersion Variance**

```python
import numpy as np
import statistics as stats
```

```python
import statistics as stats            # use by statistics

data = [10, 20 ,30 ,40, 50]

# population variance
pop_var = stats.pvariance(data)        # pvariance() -> population variance (divide by N)
print(pop_var)

# Sample variance
sam_var = stats.variance(data)         # variance() -> sample variance (divide by n-1)
print(sam_var)

```

```python
# Standard Deviation
pop_st = stats.pstdev(data)
print(pop_st)

sam_st = stats.stdev(data)
print(sam_st)
```

```python
import numpy as np                 # use by numpy

data = np.array([10, 20, 30, 40, 50])

# population variance
pop_var = np.var(data)
print(pop_var)

# sample variance
sam_var = np.var(data, ddof=1)
print(sam_var)
```

```python
# Standard deviation
pop_sd = np.std(data)
print(pop_sd)

sam_sd = np.std(data, ddof=1)
print(sam_sd)
```

### **Use by Manual Formula**

```python
data = [ 5, 10, 15, 220, 25, 300, 35, 40, 45]

mean = sum(data)/ len(data)         # mean
mean
```

```python
# Square deviation

sq_dev = [(x - mean)**2 for x in data]
sq_dev
```

```python
# Population variance
pop_var = sum(sq_dev)/ len(data)
print(format(round(pop_var,2)))        # format(round) -> decrease decimal value
```

```python
# sample variance
sam_var = sum(sq_dev)/ (len(data)- 1)     # divide by n-1
print(format(round(sam_var,2)))
```
