# **Measure Of Central Tendency**

```python
import statistics as stats
```

### **1. Mean (Average)**
- Formula: Mean = sum of all values / No. of values
- useful jab data evenly distributed ho

```python
data = [15, 20 ,25, 20, 30, 35, 20, 40, 25 ,20]

mean_value = stats.mean(data)
mean_value
```

### **2. Median (Middle Value)**
- Arrange data in ascending order
- if odd no. values -> middle value median
- if even no. values -> Average of two middle values is median

```python
median_value = stats.median(data)
median_value
```

### **3. Mode (Most Frequent Value)**
- values repeat in many times

```python
mode_value = stats.mode(data)
mode_value
```
